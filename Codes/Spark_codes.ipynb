###EDA Section

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


trainDF = train.toPandas()
testDF = test.toPandas()
```

```
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

#Figure 1
#Plot SalePrice
train = (train
         .withColumn("SalePriceTmp", col("SalePrice").cast(DoubleType()))
         .drop("SalePrice")
         .withColumnRenamed("SalePriceTmp", "SalePrice")
)
display(train)
```

```
numeric_cols = []
for colname in trainDF.columns:
  try:
    trainDF.loc[:, colname] = trainDF[colname].astype(np.int32)
  except:
    numeric_cols.append(colname)

trainDF.drop(columns=["Id"], inplace=True)

#Figure 2
#plot the correlations between features and label.
featuresCorrelation = trainDF.corr()
fig, ax = plt.subplots(figsize=(40,25))
ax = sns.heatmap(featuresCorrelation, cmap="YlGnBu" ,square=True)
ax.set_xticklabels([x.get_text() for x in ax.get_xticklabels()], rotation=90)
display(fig)
```

```
#Figure 3 sale price ~ overall quality (Rates the overall material and finish of the house)
#plot the correlation between SalePrice and Overall Quallity

sale_overallQ = trainDF[['OverallQual', 'SalePrice']]
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.boxplot(x='OverallQual', y="SalePrice", data=sale_overallQ, color= 'g')
display(fig)
```
